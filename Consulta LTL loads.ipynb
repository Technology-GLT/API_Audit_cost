```python
import time
from datetime import datetime, timedelta
today = datetime.today() #Dia actual
skip = 0
limit = 29999
url = 'https://api-audit-costs.charliebot.ai/api/v1/quote'
headers = {"apikey": "2acbe295cc203e15"}
```

```python
full_data = pd.DataFrame()
for i in range(0, len(df)-1):#
    try:
        params ={"skip": skip,
                "limit": limit,
                "LoadNumber": "{}".format(df['Load: Load Number'][i])}
        r = requests.get(url, headers=headers, params = params)
        response = r.json()
        data = pd.json_normalize(response)
        all_data = data[['load_number','carrier', 'price', 'carrier_price', 'rules_price']].copy()
        full_data = pd.concat([full_data, all_data])
    except:
        print("Load {} not found".format(df['Load: Load Number'][i]))
```
